```python
# una vez que tengamos nuestro DataFrame preparado con todas las columnas que queremos vamos a crear una función que realice una exploración inicial del conjunto de datos
def eda(dataframe):
    """
    Realiza un análisis exploratorio básico de un DataFrame, mostrando información sobre duplicados,
    valores nulos, tipos de datos, valores únicos para columnas categóricas y estadísticas descriptivas
    para columnas categóricas y numéricas

    Parámetros:
    - dataframe (DataFrame): El DataFrame que se va a explorar.
    
    Returns: 
    No devuelve nada directamente, pero imprime en la consola la información exploratoria.
    """
    
    print(f"Los duplicados que tenemos en el conjunto de datos son: {dataframe.duplicated().sum()}")
    print("\n ..................... \n")
    
    print(f"el numero de filas y columnas del df son : {df.shape}")
    print("\n ..................... \n")
    
    # generamos un DataFrame para los valores nulos
    print("Los nulos que tenemos en el conjunto de datos son:")
    df_nulos = pd.DataFrame(dataframe.isnull().sum() / dataframe.shape[0] * 100, columns = ["%_nulos"])
    display(df_nulos[df_nulos["%_nulos"] > 0])
    
    # generamos un DataFrame para tipos de datos
    print("\n ..................... \n")
    print(f"Los tipos de las columnas son:")
    display(pd.DataFrame(dataframe.dtypes, columns = ["tipo_dato"]))

    # principales estadisticos de todas las columnas
    print("\n ..................... \n")

    # estadisticos columnas numericas
    print(f"estadisticos columnas numericas")
    display(df.describe().T)
    print("\n ..................... \n")

    # estadisticos columnas categoricas
    print(f"estadisticos columnas categoricas")
    display(df.describe(include = "O").T)
    
    print("\n ..................... \n")
    print("Los valores que tenemos para las columnas categóricas son: ")
    dataframe_categoricas = dataframe.select_dtypes(include = "O")
    
    for col in dataframe_categoricas.columns:
        print(f"La columna {col.upper()} tiene las siguientes valore únicos:")
        display(pd.DataFrame(dataframe[col].value_counts()))    
    
```

```python
escala_precios = {
    1: 1,
    2: 2,
    3: 3,
    4: 4
}

def escala_precio(precio):
  """Convierte un valor de precio a una escala de 1 a 4."""
  if pd.isnull(precio):
    return 0
  return escala_precios[len(precio)]

df["Price"] = df["Price"].map(escala_precio)

```

```python


def extract_city(valor):
#Extracts the city name from a single value (handles strings and non-strings).
    if pd.isnull(valor):  # Check for NaN
        return None  # Return None or a specific valor for missing data
    try:
        return valor.split(",")[0]  # Split and get first element (assuming city is first)
    except:
        return valor  # Return original value if unable to split (e.g., no comma)



```

```python
df["City"] = df["Location"].apply(extract_city)
```

```python

def extract_country(valor):
#Extracts the city name from a single value (handles strings and non-strings).
    if pd.isnull(valor):  # Check for NaN
        return None  # Return None or a specific valor for missing data
    try:
        return valor.split(",")[1]  # Split and get first element (assuming city is first)
    except:
        return valor  # Return original value if unable to split (e.g., no comma)
```

```python
df["Country"] = df["Location"].apply(extract_country)
```

```python
from selenium import webdriver  
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.support.ui import Select 
from time import sleep
```

```python
def get_url(driver, accept_cookies = False): 

    url= "https://www.google.es/"

    driver.get(url)
    print("url loaded")
    # driver.maximize_window()
    if accept_cookies:
        driver.find_element("css selector", "#L2AGLb > div").click()
        sleep(3)
    return driver
    
```

```python

## Extraccion datos Rotten Tomatoes

url= "https://www.google.es/"


options = webdriver.FirefoxOptions()
options.page_load_strategy = "eager"
driver = webdriver.Firefox(options)

driver.get(url)

# Aceptar las cookies de GOOGLE
driver.find_element("css selector", "#L2AGLb > div").click()
sleep(3)

puntuacion_restaurantes = []

start = 201
stop = len(df)
counter = start

for element in list_busqueda[start:stop+1]:
     
    for restaurante,ciudad in element.items():
         
        busqueda = f"Restaurante: {restaurante} - ciudad: {ciudad}"
        print(busqueda)
        
        try:
            buscador = driver.find_element("css selector", "#APjFqb")
            buscador.clear()
            buscador.send_keys(busqueda, Keys.ENTER)
            sleep(3)
            
            try: 
                print(f'Buscando informacion para {restaurante} en {ciudad}')

                puntuacion_restaurantes.append((restaurante, driver.find_element("css selector","#rhs > div.kp-wholepage-osrp > div.wPNfjb > div > div > div:nth-child(2) > div > div > div.nwVKo > div.loJjTe > div > span.Aq14fc").text))
            except: 
                print("No existe puntuacion")
        except Exception as e:
            print(e)
    
    if counter % 100 == 0:
        df_seguridad =  pd.DataFrame(puntuacion_restaurantes, columns=["restaurante", "puntuacion"])
        df_seguridad.to_csv(f"df_seguridad{counter}.csv")
        driver.close()
        driver = webdriver.Firefox(options)
        driver = get_url(driver, True)
    counter += 1



df_restaurantes_google = pd.DataFrame(puntuacion_restaurantes, columns=["restaurante", "puntuacion"])
df_restaurantes_google.to_csv(f"puntuacion_google.csv")
```
